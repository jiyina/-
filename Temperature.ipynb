```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

df = pd.read_csv("observed-timeseries-of-a.xlsx - observed-timeseries-of-a.csv")

df = df.rename(columns={"Category": "Year","Average Mean Surface Air Temperature (Annual Mean °C)": "Temperature"})

X = df[["Year"]]
y = df["Temperature"]



model = LinearRegression()



##
model.fit(X, y)


y_pred = model.predict(X)

#r val
print("Temperature Regression:")
print(f"y = {model.coef_[0]:.2f}x + {model.intercept_:.2f}")
print("R² =", round(r2_score(y, y_pred), 3))
print()


#plot
plt.scatter(X, y)
plt.plot(X, y_pred)


plt.xlabel("Year")
plt.ylabel("Temperature")
plt.title("Avg Temp vs Year")
plt.show()


```
