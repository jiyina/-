```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

df = pd.read_csv("Copy of Copy of cassava & corn 1961-2023 - Sheet1 (1).csv")


##i thinkkk
cassava_df = df[df["Item"] == "Cassava, fresh"][["Year", "Value"]].dropna()
corn_df = df[df["Item"] == "Maize (corn)"][["Year", "Value"]].dropna()


X_cassava = cassava_df[["Year"]]
X_corn = corn_df[["Year"]]
y_cassava = cassava_df["Value"]
y_corn = corn_df["Value"]

model_cassava = LinearRegression()
model_corn = LinearRegression()


##
model_cassava.fit(X_cassava, y_cassava)
model_corn.fit(X_corn, y_corn)

cassava_pred = model_cassava.predict(X_cassava)
corn_pred = model_corn.predict(X_corn)

print("Cassava Regression:")
print(f"y = {model_cassava.coef_[0]:.2f}x + {model_cassava.intercept_:.2f}")
print("R² =", round(r2_score(y_cassava, cassava_pred), 3))
print()

print("Corn Regression:")
print(f"y = {model_corn.coef_[0]:.2f}x + {model_corn.intercept_:.2f}")
print("R² =", round(r2_score(y_corn, corn_pred), 3))

#plot
plt.scatter(X_cassava, y_cassava)
plt.plot(X_cassava, cassava_pred)

plt.scatter(X_corn, y_corn)
plt.plot(X_corn, corn_pred)

plt.xlabel("Year")
plt.ylabel("Production")
plt.title("Linear Regression: Cassava & Corn (1961–2023)")
plt.show()
```
